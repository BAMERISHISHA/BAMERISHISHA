**DATA ANALYSIS WORLD CARBON DIOXIDE EMISSIONS**

```python
import pandas as pd # a python library for data analysis
from matplotlib import pyplot as plt # library for data visualization, and helps in graphical plotting 
import plotly.express as px #plotly.express library is used for creating figures in python
```

```python
carbon = pd.read_csv("CARBON.csv") #reading the dataset taken from world bank 
#emissions measured in metric tons per capita 
carbon 
```

REMOVING NULL VALUES 


```python
carbon1 = carbon.dropna() # removing the missing values 
carbon1 
```

```python
info = carbon1.info() #prints information about the dataframe, contains number of columns, column  labels, column datatypes, number of cells in each column)
```

```python
# finding the country with the highest carbon dioxide emmisions 
most_polluted_2018 = carbon1.sort_values(by = '2018', ascending=False) # sort the data in decending order 
top10 = most_polluted_2018.head(10) #  getting top 10 highest carbon dioxide emitters in the world 
print(top10)
```

```python
bar = px.bar(x='Country Name', y='2018', data_frame=top10.head(10), title="Top 10 Countries with highest Co2 Emissions in 2018") 
#assinging country name to the x axis and the CO2 emissions from 2018 of the top 10 countires on the y axis
bar
```

```python
# finding the country with the highest carbon dioxide emmisions 
most_polluted_2019 = carbon1.sort_values(by = '2019', ascending=False)
top10 = most_polluted_2019.head(10) #  getting top 10 highest carbon dioxide emitters in the world in 2019
print(top10)
```

```python
pie = px.pie(top10, names = 'Country Name', values = '2019') #Creating a pie chart to show the percentage of top 10 countries with highest co2 emissions in the world in 2019
pie
```
